```python
#2 Tampilkan image dalam channel Red-Blue dan Green-Blue saja!
# Load gambar
img = cv.imread('/content/drive/MyDrive/PCVK/kucing oren.jpg')
# Konversi gambar dari BGR ke RGB
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
# Buat variable RGB untuk dilakukan split warna pada gambar
R,G,B = cv.split(img2)
# Method zeros digunakan untuk membuat array kosong
zeros = np.zeros(img2.shape[:2], dtype="uint8")
# Tampilkan warna merah hijau
red_blue = cv.merge([R,zeros,B])
plt.imshow(red_blue)
```

```python
green_blue = cv.merge([zeros,G,B])
plt.imshow(green_blue)
```

```python
#3 Tampilkan image baris ke 20-115, kolom 25-120!
img_kitten3 = cv.imread('/content/drive/MyDrive/PCVK/kucing oren.jpg')
img_kitten4 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)
crop = img_kitten4[20:115, 25:120]
plt.imshow(crop)
```

```python
#4 Tampilkan image baris ke 5-30, semua kolom, channel Red saja!
red = cv.merge([R,zeros,zeros])
y=5
x=0
h=30
w=red.shape[1]
crop = red[y:h, x:w]
plt.imshow(crop)
```

```python
#5 Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. Disarankan menggunakan bilangan acak/random!
img5 = cv.imread('/content/drive/MyDrive/PCVK/kucing oren.jpg')
img5 = cv.cvtColor(img5,cv.COLOR_BGR2RGB)
cv.rectangle(img5, pt1=(300,300), pt2=(40,170), color=(0,255,255), thickness=10)
cv.rectangle(img5, pt1=(500,100), pt2=(400,200), color=(0,255,0),thickness=10)
cv.rectangle(img5, pt1=(400,400), pt2=(500,300), color=(255,255,255),thickness=10)
cv.rectangle(img5, pt1=(100,200), pt2=(300,400), color=(255,0,0),thickness=10)
cv.rectangle(img5, pt1=(150,300), pt2=(300,150), color=(0,0,255),thickness=10)
plt.imshow(img5)
```
